```python
from pyspark.sql import SparkSession
```

```python
from pyspark import SparkContext
```

```python
spark = SparkSession.builder.appName("PySpark WordCount Test").master("local[*]").getOrCreate()
```

```python
print(type(spark))
```

```python
data = ["hello spark", "hello docker", "hello worker"]
```

```python
rdd = spark.sparkContext.textFile("/opt/spark/works/sample.text")
```

```python
counts = rdd.flatMap(lambda x: x.split(" ")).map(lambda word: (word, 1)).reduceByKey(lambda a, b: a + b)
```

```python
for word, count in counts.collect():
    print(f"{word}: {count}")
```

```python
df = counts.toDF(["word", "count"])
```

```python
print(df)
```

```python
df.show()
```

```python
df.coalesce(1).write.mode("overwrite").csv("/opt/spark/works/output/wordcount", header =True)
```

```python
df.coalesce(1).write.mode("overwrite").json("/opt/spark/output/wordcount_json")
```
