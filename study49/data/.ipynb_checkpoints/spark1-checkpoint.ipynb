```python
import os
import sys

# 현재 주피터가 실행 중인 uv 환경의 파이썬 경로를 명시
os.environ['PYSPARK_PYTHON'] = sys.executable
os.environ['PYSPARK_DRIVER_PYTHON'] = sys.executable
```

```python
from pyspark.sql import SparkSession

# 세션 생성
spark = SparkSession.builder \
    .appName("DebugApp") \
    .master("local[*]") \
    .getOrCreate()

sc = spark.sparkContext

# 아주 간단한 데이터로 테스트
test_rdd = sc.parallelize(["apple banana", "apple orange"])
try:
    # 1단계: 분리
    words = test_rdd.flatMap(lambda x: x.split(" "))
    # 2단계: 매핑 (여기서 에러가 난다면 파이썬 경로 문제입니다)
    pairs = words.map(lambda x: (x, 1))
    # 3단계: 확인
    print(pairs.take(5))
except Exception as e:
    print("에러 발생:", e)
```
